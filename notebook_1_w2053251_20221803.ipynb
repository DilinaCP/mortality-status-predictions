

*   Using IQR statistical method to find outliers to find outliers
*   From tutorial 1&2 (Week 1), Page 31



```python
# Function to identify outliers in a DataFrame column using the IQR method
def find_outliers_IQR(df):
    q1 = df.quantile(0.25)  # Calculate 1st quartile (25th percentile)
    q3 = df.quantile(0.75)  # Calculate 3rd quartile (75th percentile)
    IQR = q3 - q1           # Compute interquartile range
    # Find values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR as outliers
    outliers = df[((df < (q1 - 1.5 * IQR)) | (df > (q3 + 1.5 * IQR)))]
    return outliers
```



*   Calculate IQR, and use the values to find the outliers in the data frame
*   From tutorial 1&2 (Week 1), Page 31



```python
import pandas as pd

# Find and print outliers in the 'Age' column using the IQR method
outliers = find_outliers_IQR(data['Age'])
print(outliers)

# Find outliers in the 'Tumor_Size' column, set display option to show all rows, then print them
outliers = find_outliers_IQR(data['Tumor_Size'])
pd.set_option('display.max_rows', None)
print(outliers)

# Find and print outliers in the 'Survival_Months' column using the IQR method
outliers = find_outliers_IQR(data['Survival_Months'])
print(outliers)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Clean_Coursework_Dataset1.csv')

# Define numerical and categorical features
numerical_features = ['Age', 'Tumor_Size', 'Regional_Node_Examined','Reginol_Node_Positive',
                      'Estrogen_Status', 'Progesterone_Status', 'A_Stage']

categorical_features = ['T_Stage', 'N_Stage', '6th_Stage', 'Differentiated', 'Grade']

# Encode categorical features
label_encoders = {}
for col in categorical_features:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])  # Convert categorical to numeric
    label_encoders[col] = le  # Save encoder for later use (if needed)

# Combine numerical & encoded categorical features
all_features = numerical_features + categorical_features

# Fill missing values (if any)
df[all_features] = df[all_features].fillna(df[all_features].median())

# Apply Min-Max Scaling
scaler = MinMaxScaler()
df[all_features] = scaler.fit_transform(df[all_features])

# Define features (X) and targets (y)
X = df[all_features]
y_class = df['Mortality_Status']  # Classification Target
y_reg = df['Survival_Months']     # Regression Target

# Split dataset into training and testing sets
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X, y_class, test_size=0.2, random_state=42)
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_reg, test_size=0.2, random_state=42)

# Print dataset shapes
print("Classification Training Set: ", X_train_class.shape, y_train_class.shape)
print("Classification Test Set: ", X_test_class.shape, y_test_class.shape)
print("Regression Training Set: ", X_train_reg.shape, y_train_reg.shape)
print("Regression Test Set: ", X_test_reg.shape, y_test_reg.shape)

# Create dataframes for saving
classification_df = pd.concat([X, y_class], axis=1)
regression_df = pd.concat([X, y_reg], axis=1)

# Save the processed datasets permanently in Google Drive
classification_df.to_csv('/content/drive/MyDrive/Colab Notebooks/classification_data.csv', index=False)
regression_df.to_csv('/content/drive/MyDrive/Colab Notebooks/regression_data.csv', index=False)

```
